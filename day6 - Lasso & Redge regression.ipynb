```python
import pandas as pd
df=pd.read_csv('Top Indian Places to Visit.csv',index_col='Unnamed: 0')
import warnings
warnings.filterwarnings("ignore")
```

```python
df
```

```python
df.nunique()
```

```python
col=['Type','Google review rating','Number of google review in lakhs','Entrance Fee in INR']
dff=df[col]
dff
```

```python
dff = dff[dff['Entrance Fee in INR'] != 0]
dff
```

```python
dff = pd.get_dummies(dff,dtype=int)
dff
```

```python
x=dff.drop(columns=['Entrance Fee in INR'])
y=dff['Entrance Fee in INR']
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3)
```

```python
reg=LinearRegression()
reg.fit(x_train,y_train)
```

```python
reg.score(x_train,y_train)
```

```python
reg.score(x_test,y_test)
```

# Lasso reg

```python
from sklearn.linear_model import Lasso
reg2= Lasso(alpha=0,max_iter=100,tol=0.01)
reg2.fit(x_train,y_train)
```

```python
reg2.score(x_train,y_train)
```

```python
reg2.score(x_test,y_test)
```

# Ridge regression

```python
from sklearn.linear_model import Ridge
reg3= Ridge(alpha=15,max_iter=100,tol=0.1)
reg3.fit(x_train,y_train)
```

```python
reg3.score(x_train,y_train)
```

```python
reg3.score(x_test,y_test)
```
