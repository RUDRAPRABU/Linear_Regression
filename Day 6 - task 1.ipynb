```python
import pandas as pd
df=pd.read_csv('NY-House-Dataset.csv')
```

```python
df
```

```python
df.nunique()
```

```python
df.corr()
```

```python
col=['TYPE','BEDS','BATH','PROPERTYSQFT','ADMINISTRATIVE_AREA_LEVEL_2','LOCALITY','SUBLOCALITY','STREET_NAME','PRICE']
dff=df[col]
dff
```

```python
dff.isna().sum()
```

```python
dff=pd.get_dummies(dff,dtype=int)
dff
```

```python
x=dff.drop(columns=['PRICE'])
y=dff['PRICE']
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.15)
```

```python
from sklearn.linear_model import LinearRegression
reg=LinearRegression()
reg.fit(xtrain,ytrain)
```

```python
reg.score(xtrain,ytrain)
```

```python
reg.score(xtest,ytest)
```

```python
from sklearn.linear_model import LogisticRegression
reg1=LogisticRegression()
reg1.fit(xtrain,ytrain)
```

```python
print(reg1.score(xtrain,ytrain))
print(reg1.score(xtest,ytest))
```

```python
from sklearn.linear_model import Lasso
reg2 = Lasso(alpha=75,max_iter=100,tol=0.1)
reg2.fit(xtrain,ytrain)
```

```python
print(reg2.score(xtrain,ytrain))
print(reg2.score(xtest,ytest))
```

```python
from sklearn.linear_model import Ridge
reg3= Ridge(alpha=75,max_iter=100,tol=0.1)
reg3.fit(xtrain,ytrain)
```

```python
print(reg3.score(xtrain,ytrain))
print(reg3.score(xtest,ytest))
```
