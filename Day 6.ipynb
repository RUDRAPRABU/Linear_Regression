```python
import warnings
warnings.filterwarnings("ignore")
import numpy as np 
import pandas as pd
```

```python
dff=pd.read_csv('Melbourne_housing_FULL.csv')
dff
```

```python
col=['Suburb','Rooms','Type','Method','SellerG','Regionname','Propertycount','Distance','CouncilArea','Bedroom2','Bathroom','Car','Landsize','BuildingArea','Price']
df=dff[col]
df
```

```python
df.dropna(inplace=True)
df
```

```python
dfnew=pd.get_dummies(df,dtype='int')
dfnew
```

```python
x=dfnew.drop(columns=['Price'])
y=dfnew['Price']
x
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.15)
```

```python
print(y_test)
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(x_train,y_train)
```

```python
model.score(x_train,y_train)
```

```python
model.score(x_test,y_test)
```

```python
from sklearn.linear_model import Lasso
model2= Lasso(alpha=50,max_iter=100,tol=0.1)
model2.fit(x_train,y_train)
```

```python
model2.score(x_train,y_train)
```

```python
model2.score(x_test,y_test)
```
